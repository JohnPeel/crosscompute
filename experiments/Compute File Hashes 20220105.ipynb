```python
from os.path import expanduser
configuration_path = expanduser('~/Projects/crosscompute-examples/automate.yml')
configuration_path
```

```python
from crosscompute.constants import MODE_NAMES
from crosscompute.routines.automation import Automation
automation = Automation.load(configuration_path)
```

```python
automation.automation_definitions[0]['path']
```

```python
self = automation
```

```python
def get_paths_from_folder(folder):
    paths = set()
    for automation_definition in self.automation_definitions:
        for mode_name in MODE_NAMES:
            mode_configuration = automation_definition.get(mode_name, {})
            for variable_definition in mode_configuration.get('variables', []):
                if 'path' in variable_definition.get('configuration', {}):
                    paths.add(join(folder, variable_definition['configuration']['path']))
                paths.add(join(folder, variable_definition['path']))
    return paths
```

```python
from crosscompute.routines.configuration import load_configuration
```

```python
from os.path import join

paths = set()
for automation_definition in self.automation_definitions:
    automation_folder = automation_definition['folder']
    configuration = load_configuration(automation_definition['path'])

    for mode_name in MODE_NAMES:
        mode_configuration = automation_definition.get(mode_name, {})
        for template_definition in mode_configuration.get('templates', []):            
            if 'path' not in template_definition:
                continue
            paths.add(join(automation_folder, template_definition['path']))
    
    # TODO: Watch only on view
    for batch_definition in automation_definition.get('batches', []):
        paths.update(get_paths_from_folder(join(automation_folder, batch_definition['folder'])))
            
    for batch_definition in configuration.get('batches', []):
        if 'path' not in batch_definition.get('configuration', {}):
            continue
        paths.add(join(automation_folder, batch_definition['configuration']['path']))

    display_configuration = automation_definition.get('display', {})
    for style_definition in display_configuration.get('styles', []):
        if 'path' not in style_definition:
            continue
        paths.add(join(automation_folder, style_definition['path']))

    paths.add(automation_definition['path'])
paths
```
